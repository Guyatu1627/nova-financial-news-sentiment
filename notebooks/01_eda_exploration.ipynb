```python
import sys
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Fix path to see the scripts folder
sys.path.append(os.path.abspath('../'))
from scripts.data_loader import DataLoader

# Initialize
loader = DataLoader('../data/raw_analyst_ratings.csv')
loader.load_data()
loader.clean_data()
```

```python
# Get the stats
stats = loader.get_task_1_stats()

# --- Diagram 1: Publication Frequency ---
plt.figure(figsize=(10, 5))
# We sort by day of week for a professional look
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
stats['publication_days'].reindex(day_order).plot(kind='bar', color='#2980b9')
plt.title("News Volume by Day of the Week")
plt.ylabel("Number of Articles")
plt.xticks(rotation=45)
plt.show()

# --- Diagram 2: Headline Lengths ---
plt.figure(figsize=(10, 5))
sns.histplot(stats['headline_lengths'], bins=100, kde=True, color='#27ae60')
plt.title("Distribution of Headline Lengths (Character Count)")
plt.xlabel("Length of Headline")
plt.xlim(0, 200) # Most headlines are under 200 chars
plt.show()
```
