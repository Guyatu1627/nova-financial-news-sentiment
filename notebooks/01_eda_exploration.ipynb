```python
import sys
import os
import matplotlib.pyplot as plt
import seaborn as sns

sys.path.append(os.path.abspath('../'))
from scripts.data_loader import DataLoader

loader = DataLoader('../data/raw_analyst_ratings.csv')
df = loader.load_data()
df = loader.clean_data()
```

```python
# Calculate length of headlines
df['headline_len'] = df['headline'].str.len()

plt.figure(figsize=(10, 6))
sns.histplot(df['headline_len'], bins=100, color='skyblue', kde=True)
plt.title("Distribution of Headline Lengths")
plt.xlabel("Character Count")
plt.ylabel("Frequency")
plt.show()
```

```python
# Extract day of week
df['day_of_week'] = df['date'].dt.day_name()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

plt.figure(figsize=(10, 6))
df['day_of_week'].value_counts().reindex(day_order).plot(kind='bar', color='orange')
plt.title("News Volume by Day of the Week")
plt.ylabel("Article Count")
plt.xticks(rotation=45)
plt.show()
```
