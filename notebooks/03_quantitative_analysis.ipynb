```python
import sys
import os
import pandas as pd
import matplotlib.pyplot as plt

sys.path.append(os.path.abspath('../'))
from scripts.analyzer import FinancialAnalyzer

# Load your AAPL data
# Note: Ensure you have AAPL.csv in your /data folder
try:
    stock_df = pd.read_csv('../data/AAPL.csv')
    
    # Run the indicators
    analyzer = FinancialAnalyzer()
    stock_df = analyzer.calculate_technical_indicators(stock_df)

    # Plot the results
    plt.figure(figsize=(12, 6))
    plt.plot(stock_df['Date'], stock_df['Close'], label='AAPL Price', alpha=0.5)
    plt.plot(stock_df['Date'], stock_df['SMA_7'], label='7-Day SMA (Fast)', color='orange')
    plt.plot(stock_df['Date'], stock_df['SMA_21'], label='21-Day SMA (Slow)', color='red')
    plt.title("AAPL Technical Indicators")
    plt.legend()
    plt.show()

    print("Indicators Calculated!")
    print(stock_df[['Date', 'Close', 'Daily_Return']].tail())

except FileNotFoundError:
    print("❌ Error: AAPL.csv not found in the /data folder.")
```
